This notebook demonstrates professional-level data visualizations using the `seaborn` library on the famous `tips` dataset.

We will cover various plot types to analyze relationships and distributions including scatter plots, bar plots, violin plots, and more.


```python

# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Load the built-in 'tips' dataset from seaborn
tips_data = sns.load_dataset("tips")

# Display first 5 rows of the dataset to understand its structure
tips_data.head()
```

```python
# Set figure size for plots
plt.figure(figsize=(8, 6))
```

<h2 style='text-align:center'>Scatter Plot</h2>

```python

# ➤ Scatter plot of total bill vs tip, colored by time and sized by party size

plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips_data, x="total_bill", y="tip", hue="time", size="size", palette="deep")
plt.title("Scatter Plot: Total Bill vs Tip (Grouped by Time and Size)")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip Amount ($)")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>Line Plot</h2>

```python

# ➤ Line plot showing trend of total bill with party size, differentiated by gender

plt.figure(figsize=(8, 6))
sns.lineplot(data=tips_data, x='size', y='total_bill', hue='sex', marker='o')
plt.title("Line Plot: Total Bill by Party Size (Grouped by Gender)")
plt.xlabel("Party Size")
plt.ylabel("Total Bill ($)")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>Bar Plot</h2>

```python

# ➤ Bar plot of average total bill per day, segmented by gender

plt.figure(figsize=(8, 6))
sns.barplot(data=tips_data, x='day', y='total_bill', hue='sex', palette='muted')
plt.title("Bar Plot: Average Total Bill by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>Box Plot</h2>

```python

# ➤ Box plot of tip distribution per day, grouped by smoker status

plt.figure(figsize=(8, 6))
sns.boxplot(data=tips_data, x='day', y='tip', hue='smoker', palette='Set2')
plt.title("Box Plot: Tip Distribution by Day and Smoking Status")
plt.xlabel("Day of the Week")
plt.ylabel("Tip Amount ($)")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>Violin Plot</h2>

```python

# ➤ Violin plot to show total bill distribution across day and time

plt.figure(figsize=(8, 6))
sns.violinplot(data=tips_data, x='day', y='total_bill', hue='time', split=True, palette='pastel')
plt.title("Violin Plot: Total Bill by Day and Time")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>Count Plot</h2>

```python

# ➤ Count plot showing number of observations per day, grouped by smoker

plt.figure(figsize=(8, 6))
sns.countplot(data=tips_data, x='day', hue='smoker', palette='dark')
plt.title("Count Plot: Frequency of Days by Smoker Status")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Customers")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>Regression Plot</h2>

```python

# ➤ Regression line showing trend between total bill and tip

plt.figure(figsize=(8, 6))
sns.regplot(data=tips_data, x='total_bill', y='tip',
            scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title("Regression Plot: Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip Amount ($)")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>Histogram Plot</h2>

```python

# ➤ Histogram with KDE for total bill distribution

plt.figure(figsize=(8, 6))
sns.histplot(data=tips_data, x='total_bill', bins=20, kde=True, color='blue')
plt.title("Histogram: Total Bill Distribution with KDE")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>Strip Plot</h2>

```python

# ➤ Strip plot of tips per day by gender

plt.figure(figsize=(8, 6))
sns.stripplot(data=tips_data, x='day', y='tip', hue='sex', jitter=True, palette='Set1')
plt.title("Strip Plot: Tips by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Tip Amount ($)")
plt.grid(True)
plt.show()
```

<h2 style='text-align:center'>KDE Plot</h2>

```python

# ➤ KDE (Kernel Density Estimation) plot of total bill, grouped by gender

plt.figure(figsize=(8, 6))
sns.kdeplot(data=tips_data, x='total_bill', hue='sex', fill=True, palette='tab10')
plt.title("KDE Plot: Total Bill Density by Gender")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
